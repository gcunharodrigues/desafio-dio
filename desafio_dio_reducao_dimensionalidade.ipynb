```python
import cv2
import matplotlib.pyplot as plt

def rgb_to_grayscale(image_path):
  """
  Transforms a color image to grayscale using OpenCV.

  Args:
    image_path: The path to the color image file.

  Returns:
    A grayscale and black/white image as a NumPy array.
    Returns None if the image cannot be loaded.
    Also displays the original, grayscale  and black/white images using matplotlib.
  """
  try:
    # Load the color image using OpenCV
    img = cv2.imread(image_path)
    if img is None:
      print(f"Error: Could not load image at {image_path}")
      return None

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, bw_img = cv2.threshold(gray_img, 160, 255, cv2.THRESH_BINARY_INV)

    # Display the original and grayscale images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
    plt.title("Original Image")

    plt.subplot(1, 3, 2)
    plt.imshow(gray_img, cmap='gray')
    plt.title("Grayscale Image")

    plt.subplot(1, 3, 3)
    plt.imshow(bw_img, cmap='binary')
    plt.title("Black and White Image")

    plt.show()

    return gray_img
  except Exception as e:
    print(f"An error occurred: {e}")
    return None


# Example usage
grayscale_image = rgb_to_grayscale("cat-sitter-felino.jpg") # Replace 'your_image.jpg' with the actual path to your image file

if grayscale_image is not None:
  print("Conversion successful.")
  # You can further process the grayscale_image here, e.g., save it to a file
  # cv2.imwrite("grayscale_image.jpg", grayscale_image)
```
